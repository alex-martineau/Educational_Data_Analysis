# 3. Analyse pré-exploratoire<a name=3 ></a>

Tout d'abord, nous pouvons nous poser la question de la quantité d'information indiposnible par rapport à celle disponible. Pour cela, nous allons créer un fonction que nous nommerons "totalinfo" et qui permettra de déterminer le pourcentage de valeurs nulles présentes dans une base de données ciblée :

```python
def totalinfo(df):
    percentage = round(df.isna().sum().sum()/(df.shape[0]*df.shape[1]),2)*100
    print(f"La dataframe est vide à {percentage} %")
    print(f"La dataframe contient {df.shape[0]} lignes et {df.shape[1]} colonnes")
totalinfo(EdStatsData)
```

Maintenant, nous savons que 86% des valeurs de la base de données sont nulles. Autrement dit, 86% de la base de données est inexploitable. Ce chiffre est évidemment trop élevé pour que l'on puisse effectuer notre analyse exploratoire en l'état.

Il est donc indispensable de réduire drastiquement le nombre de valeurs afin que notre chiffre de 86% diminue considérablement jusqu'à atteindre un seuil acceptable. Mais par où commencer?

## 3.1 Années<a name=4 ></a>

Nous avons déjà précédemment sélectionné 11 indicateurs, ce qui déjà peut permettre de disposer d'une base de données plus complète. Nous pouvons ainsi nous intéresser à l'ensemble suivant, soit celui qui regroupe la quasi-totalité des colonnes d'EdStatsData : les années.

Ici, nous allons prendre notre formule permettant de déterminer le pourcentage d'inexploitabilité de la base de données, puis nous allons l'appliquer en fonction des années étudiées. Les résultats de ces calculs vont être affichées dans un graphique en barre dans lequel plus la barre est élevée, plus le taux est élevé, et donc moins l'année est exploitable : 

```python
sns.set(rc={'figure.figsize':(15,5)})
pd.DataFrame(EdStatsData.isna().sum()/EdStatsData.shape[0]).drop(['Country Name','Country Code','Indicator Name','Indicator Code']).plot(kind='bar', title="Pourcentage de valeurs nulles par années", figsize=[15,5])
```

Dans ce graphique, nous pouvons voir que le pourcentage de valeurs nulles demeure extrêmement élevé sur l'ensemble des années, dépassant les 80% dans la quasi-totalité des cas et allant même jusqu'à atteindre 100% en 2016.

Pour autant, 2 tendances sont à déduire : 
- Les années constituant des tranches de 5 semblent moins incomplètes que les autres
- Le niveau d'incomplétude tend à la baisse à partir du début des années 2000 jusqu'à la date du projet (2015)

Quoi qu'il en soit, l'information la plus importante de ce graphique et que l'exploitabilité de notre base de données n'est pas basée sur les années, comme nous aurions pu le supposer. Mais nous pouvons déjà établir un premier filtre en ne gardant que les années allant de 2000 à 2015.

A partir de là, le plus simple, pour pouvoir entamer notre analyse exploratoire, serait de conserver les chiffres les plus récents par pays et pour chacun de nos 11 indicateurs.

Dans cette démarche, nous allons établir la liste des 10 années retenues, à partir de laquelle nous allons créer une boucle d'extraction visant les chiffres cibles (les valeurs les plus récentes), après avoir effectué au préalable un premier travail de filtrage des pays cibles, notamment en supprimant les éléments dont la zone géographique n'est pas déterminée.

A partir de cette boucle, une nouvelle base de donnée plus épurée et que nous serons à même de pouvoir analyser efficacement, va pouvoir être créée. Il serait par ailleurs interessant de savoir à combien le pourcentage de valeurs nulles passe avec cette nouvelle base de données : 

```python
# Création de la liste "years"

years = ['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']

# Suppression des pays sans zone géographique

workCountry = EdStatsCountry[EdStatsCountry.Region.isnull() == False]
workCountry.reset_index(drop=True, inplace = True)

# Création de la nouvelle base de données "workCountry" à partir d'une boucle

for i in indicateurs:
    df_temp = EdStatsData[EdStatsData['Indicator Code'] == i].reset_index(drop=True)
    df_temp[i] = df_temp[years].apply(lambda x: x[x.notnull()].iloc[-1] if len(x[x.notnull()])>0 else np.NAN, axis=1)
    workCountry = workCountry.merge(df_temp[['Country Code', i]], on="Country Code", how='left')

# Calcul du pourcentage de valeurs nulles dans la nouvelle base de données

totalinfo(workCountry)
```

Depuis une base de données vide à 86%, nous sommes maintenant passés à 19%, ce qui est pour le moins non-négligable.

Cependant, cela ne veut en rien dire que nous pouvons nous arrêter là dans notre travail d'épuration, et donc que nous ne pouvons encore davantage faire baisser ce chiffre.

## 3.2 Indicateurs<a name=5 ></a>

Regardons maintenant à combien s'élève le pourcentage de valeurs nulles selon les 11 indicateurs retenus : 

```python
pd.DataFrame(workCountry.isna().sum()/workCountry.shape[0]).drop(['Country Code','Short Name','Region','Income Group']).drop(workCountry.columns[workCountry.columns.str.endswith(' MY')]).plot(kind='bar', title="Pourcentage de valeurs nulles par variable", figsize=[15,5], rot=45)
```

Si, dans l'ensemble, nos indicateurs présentent un pourcentage nettement plus faible, force est de constater que pour 2 d'entre elles, il atteint les 100%. Ces dernières ne contiennent donc strictement aucun chiffre à analyser et peuvent donc être définitivement rayées de la liste de nos indicateurs.

Au-delà de cela, il serait intéressant de se pencher sur les années dont sont originaires l'ensemble des valeurs de workCountry. Redistribuées en fonction des indicateurs, elles peuvent mettre en lumière le caractère récent de ces derniers : 

```python
# Retrait des indicateurs UIS.SCHBSP.2T3.WCOMPUT & UIS.SCHBSP.2T3.WINTERN & creation de la liste maxyearcount

workCountry.drop(['UIS.SCHBSP.2T3.WCOMPUT','UIS.SCHBSP.2T3.WINTERN'], axis=1, inplace=True)
del(indicateurs[5:7])
del(years[0:5])
maxyearcount = []

# Suppression des pays sans zone géographique

maxyear = EdStatsCountry[EdStatsCountry.Region.isnull() == False]
maxyear.reset_index(drop=True, inplace = True)

# Creation de la boucle

for i in indicateurs:
    df_temp = EdStatsData[EdStatsData['Indicator Code'] == i].reset_index(drop=True)
    df_temp[i] = df_temp[years].apply(lambda x: x[x.notnull()].index[-1] if len(x[x.notnull()])>0 else np.NAN, axis=1)
    maxyear = maxyear.merge(df_temp[['Country Code', i]], on="Country Code", how='left')
    maxyearcount.append(pd.DataFrame(maxyear[i].value_counts()).rename(columns={ 'count': i }))
    
# Creation de la base de données

from functools import reduce
repartition = reduce(lambda l, r: pd.merge(l, r, how='outer', left_index=True, right_index=True), maxyearcount)

# Creation du graphique

repartition.sort_index().plot(kind='bar', title="Répartition du nombre d'éléments par indicateur/année maximale", figsize=[15,5], width=1, rot=0)
```

Traduction : la grande majorité des chiffres contenus de nos indicateurs sont récents (datent de 2015), à l'exception des 2 suivants : 
- UIS.E.3-Enrolment in upper secondary education, both sexes (number) : la majorité des chiffres proviennent de l'année 2014
- IT.CMP.PCMP.P2-Personal computers (per 100 people) : la majorité des chiffres sont daté entre 2005 et 2008 et peuvent éventuellement être considérés comme anciens.

Dans l'ensemble, rien d'alarmant. Nous pouvons donc passer à la partie suivant qui se concentre sur la répartition des valeurs par région géographique.

## 3.3 Région<a name=6 ></a>

Voici donc ci-dessous un diagramme en barres représentant le pourcentage de valeurs nulles selon à la fois les indicateurs et les régions géographiques. Il permettra d'identifier, parmi ces dernières, celles qui peuvent potentiellement faire l'objet d'un procédé d'élimination :

```python
workCountry.groupby('Region').apply(lambda x: x.isnull().mean()).drop(['Country Code','Short Name','Region','Income Group'], axis=1).T.plot(kind='bar', title="Pourcentage de valeurs nulles par variable/région", figsize=[15,5], rot=45)
```

Le premier constat à tirer de ce graphique est que les indicateur qui suivent présentent un pourcentage de valeurs nulles qui, une fois répartis par région, sont nettement plus élevés que les autres : 
- SE.TER.PRIV.ZS (Percentage of enrolment in tertiary education in private institutions (%))
- SE.SEC.PRIV.ZS (Percentage of enrolment in secondary education in private institutions (%))

Le deuxième constat est que la région East Asia & Pacific regroupe un taux de remplissage de valeurs parmi les plus faibles, et ce en particulier concernant les 5 premiers indicateurs du graphiques.

Nous allons donc effectuer 2 opérations. Tout d'abord, nous allons supprimer les 2 indicateurs mentionnés ci-dessus. Ensuite, nous allons procéder à un filtrage massif des pays afin de ne garder que ceux dont les chiffres sont les plus complets.

Pour cela, rien de plus simple, nous allons créer une nouvelle colonne dans laquelle sera indiqué le nombre total de valeurs nulles par ligne. Sachant que nous opérons notre analyse sur 8 indicateurs, nous allons ensuite filtrer ladite colonne pour pouvoir conserver uniquement les pays dont le nombre de valeurs nulles ne dépasse pas 5 : 

```python
# Retrait des indicateurs SE.TER.PRIV.ZS & SE.SEC.PRIV.ZS

workCountry.drop(['SE.TER.PRIV.ZS','SE.SEC.PRIV.ZS'], axis=1, inplace=True)
del(indicateurs[3:5])

# Création de la colonne & suppression des pays aux données insuffisantes

workCountry['NullCount'] = workCountry.isnull().sum(axis=1)
workCountry.drop(workCountry.loc[workCountry.NullCount >= 5].index, inplace=True)
totalinfo(workCountry)
```

Le pourcentage de valeurs nulles est maintenant passée à 2%, ce qui constitue désormais un seuil plus qu'acceptable. De plus, d'après le graphique ci-dessous, ce chiffre atteint une valeur maximale de 12,5% (elle concerne le taux d'inexploitabilité de l'indicateur SE.TER.ENRL pour les pays d'Asie de l'Est & du Pacifique) ce qui constitue là aussi un seuil acceptable :

```python
workCountry.groupby('Region').apply(lambda x: x.isnull().mean()).drop(['Country Code','Short Name','Region','Income Group','NullCount'], axis=1).drop(workCountry.columns[workCountry.columns.str.endswith(' MY')], axis=1).T.plot(kind='bar', title="Pourcentage de valeurs nulles par variable/région", figsize=[15,5], rot=0)
```

Nous pouvons désormais considérer que notre base de données est opérationnelle pour une analyse exploratoire. Cependant, avant de la débuter, nous pouvons encore supprimer certains indicateurs. En effet, nous savons maintenant que la totalité de nos indicateurs sont opérationnels. Nous pouvons donc nous permettre de nous passer de certains d'entre eux, car leurs contenus peuvent présenter des similitudes, voire des valeurs corrélées, avec d'autres.

Pour cette raison, nous allons retirer les 3 indicateurs suivants :
- SP.POP.TOTL (Population, total) : nous disposons actuellement des chiffres concernant la population par pays âgé entre 15 et 24 ans. Dans cette mesure, le chiffre de la population total ne constitue plus une donnée aussi utile.
- NY.GDP.MKTP.PP.KD (GDP, PPP (constant 2011 international $)) : nous disposons déjà d'une variable économique en la variable de "Income Group"
- UIS.E.3 (Enrolment in upper secondary education, both sexes (number)) : nous ne disposons pas des chiffres pour l'indicateur équivalent dans le secteur tertiaire. De plus, les indicateurs SE.SEC.ENRL et SE.TER.ENRL suffisent en soi à servir d'indicateur sur le plan éducatif.




```python
workCountry.drop(['SP.POP.TOTL','NY.GDP.MKTP.PP.KD','UIS.E.3'],axis=1,inplace=True)
del(indicateurs[0])
del(indicateurs[4])
del(indicateurs[5])
totalinfo(workCountry)
```

Malgré la supression de 3 indicateurs extrêmement complets (en particulier la population totale, valide à 100%), le taux de remplissage de notre base de données demeure intacte. Nous pouvons débuter dès maintenant notre analyse exploratoire.

Pour commencer, nous pouvons examiner les types de valeur de chacune de nos variables :

```python
workCountry.dtypes
```

Rien de particulier à signaler ici, si ce n'est que les valeurs des indicateurs que nous allons étudier sont bien numériques, exception faite de la variable "Income Group". Nous pouvons passer à l'étape d'après, à savoir établir une vue d'ensemble de la base de données via des outils statistiques (moyenne, médiane, écart-type, etc.) : 

```python
workCountry.describe(include='all')
```

Beaucoup d'informations significatives peuvent être extraites ici. Pour pouvoir y voir plus clair, nous allons nous intéresser aux valeurs aberrantes, ou outliers, présents dans notre base de données.

Pour ce faire, nous allons créer une fonction permettant de calculer les 1er et 3ème quartiles et donc, par extension, la limite permettant de distinguer lesdits outliers.

Nous appliquerons ensuite cette fonction sur l'ensemble de nos indicateurs tout en veillant à regrouper le nombre de ces outliers par région : 

```python
# Préparation des listes outvariable & outcountry

outvariable = []
outcountry = []

# Création de la fonction outlier

def outlier(df,col):
    Q1 = np.percentile(df[col].dropna(), 25)
    Q3 = np.percentile(df[col].dropna(), 75)
    Lim = ((Q3-Q1)*1.5)+Q3
    return df[df[col]>=Lim]

# Création de la boucle des outliers par région

for i in indicateurs : 
    outvariable.append(outlier(workCountry,i).groupby('Region').count().rename(columns={ 'Country Code': i }).iloc[:,0:1])
    outcountry.append(outlier(workCountry,i))
    
# Fusion des données outliers

from functools import reduce
variablecount = reduce(lambda l, r: pd.merge(l, r, how='outer', left_index=True, right_index=True), outvariable)
countrycount = reduce(lambda l, r: pd.concat([l, r], axis = 0), outcountry)

variablecount
```

Le tableau ci-dessus démontre un nombre non-néglieable d'outliers, à l'exception de la variable liée au pourcentage de la population utilisatrice d'Internet. En effet, aucune outlier à signaler ici. En revanche, il s'agit, dans l'ensemble, de chiffres bruts et de ce fait, l'existence d'outliers dans des variables démographiques n'est pas si surprenant.

C'est, par ailleurs, ce que démontre le tableau ci-dessous, qui calcule le nombre d'outliers par pays plus spécifiquement. Nous y voyons que plusieurs de ces pays font l'objet de valeurs aberrantes de façon récurrente : 

```python
countrycount.groupby(['Region','Short Name']).count().iloc[:,0:1]
```

Notre analyse exploratoire démontre une base de données cohérente et conforme aux réalités des enjeux dont fait l'objet le projet d'expansion d'academy. Nous sommes maintenant prêts à analyser de manière plus spécifique chacun de nos indicateurs. L'analyse univariée dont fera l'objet la partie qui suit aura pour finalité de débuter une sélection de pays répondant aux besoins de notre projet.

# 6. Sélection des pays<a name=9 ></a>

Avant de débuter notre système de notation, il est déjà possible d'effectuer un premier tri grâce à la variable "Income Group". En effet, cette dernière constitue une variable faussement qualitative, mais il n'est pas nécessaire ici de la modifier. Les valeurs qu'elle contient peuvent d'ores et déjà être considérées comme le résultat d'une notation.

De plus, dans le cadre de notre projet d'expansion, nous savons que nous visons les pays avec des valeurs économiques très spécifiques. Voilà pourquoi nous pouvons, dans une première étape, ne conserver que les pays dont la valeur de la variable "Income Group" est égale soit à "High Income: OECD" et "High Income: Non-OECD".

Dans un deuxième temps, nous procéderons à notre méthode de scoring grâce à la formule .qcut() permettant de répartir les individus selon leurs valeurs en se basant sur l'ensemble de ces dernières. Nous allons donc appliquer cette méthode sur nos 5 indicateurs restants avec un score maximal de 10 via une boucle.

Avec les scores obtenus, nous calculerons la moyenne de chaque individu et selectionnerons ceux dont la moyenne est supérieure ou égale à 7 : 

```python
# Supression des pays n'ayant pas un revenu considéré comme élevé

workCountry = workCountry[~workCountry['Income Group'].isin(['Low income','Lower middle income','Upper middle income'])]

# Création de la boucle de scoring

for i in indicateurs : 
    workCountry[f"{i}_score"] = pd.qcut(workCountry[i], 10, labels=False)

# Calcul du scoring moyen par pays
    
workCountry['score_mean'] = round((workCountry[workCountry.columns[workCountry.columns.str.endswith('_score')]].mean(axis=1)),1)

# Selection finale des pays d'après le scoring moyen

finalCountry = workCountry[workCountry['score_mean']>=7]
finalCountry
```

Nous avons enfin effectué ici notre sélection finale de pays répondant aux enjeux du projet d'expansion d'academy. Nous pouvons constater que les 10 pays finalistes sont répartis entre 3 régions géographiques : 
- L'Amérique du Nord (Etats-Unis, Canada)
- L'Eurore et l'Asie Centrale (Allemagne, France, Royaume-Uni, Pays-Bas, Suede)
- Asie de l'Est & Pacifique (Australie, Japon, Corée du Sud)

De plus, ils ont tous un point commun : ils sont tous membres de l'OCDE. Enfin, si academy souhaite développer son activité dans un pays partageant des similitudes démographiques, économiques, éducatives et technologiques avec la France, pays dans lequel il est actuellement implanté, la présence de cette dernière dans notre liste des pays finalistes peut confirmer la fiabilité de notre sélection.

A partir donc de cette liste, que pour rappel nous avons nommé finalist, nous sommes maintenant prêts à entamer la dernière partie de notre rapport, soit une analyse prévisionnelle pour ces pays via des données de projection.

# 7. Analyses Chronologique & Prévisionnelle<a name=10 ></a>

Avant de s'attaquer aux données de projection, il peut être intéressant de visualiser et de mettre en parallèle entre chacun des 10 pays la manière dont les valeurs qui nous intéressent ont évolué dans le temps jusqu'à aujourd'hui.

Ainsi, nous allons reprendre la base de données de départ, EdStatsData, et tout simplement afficher les données désirées au fil des années grâce à une boucle : 

```python
# Préparation de la boucle

finalist = list(finalCountry['Country Code'].values)
historic = list(EdStatsData.iloc[:,4:-19].columns.values)

# Création de la boucle des graphiques chronologiques

for i in indicateurs : 
    EdStatsData[(EdStatsData['Indicator Code']==i)&((EdStatsData['Country Code'].isin(finalist)))][['Country Code']+historic].set_index("Country Code")[historic].T.plot.line()
    plt.title(f"{EdStatsSeries[EdStatsSeries['Series Code']==i].iloc[0,1]}")
```
